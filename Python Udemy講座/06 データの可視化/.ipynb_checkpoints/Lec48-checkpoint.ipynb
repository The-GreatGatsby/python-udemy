```python
import numpy as np
from numpy.random import randn
import pandas as pd

# scipyの統計パッケージも使います。
from scipy import stats

# 描画のためのライブラリです。
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# ブラウザ内に画像を埋め込むための命令です
%matplotlib inline
```

ヒストグラムの説明は以下を参照してください。
https://ja.wikipedia.org/wiki/%E3%83%92%E3%82%B9%E3%83%88%E3%82%B0%E3%83%A9%E3%83%A0

```python
# 正規分布に従う乱数を作ります。
dataset1 = randn(100)
# bins=10がデフォルトです。
plt.hist(dataset1)
```

```python
# もう一つ作ります。
dataset2 = randn(800)
# 色を変えてみましょう。
plt.hist(dataset2,color='indianred')
```

```python
plt.hist(dataset1, normed=True)
# normed=Trueで縦軸を度数ではなく、比率を表示。全部足すと1．
```

```python
#####################################################################

# alpha値で、図を透けさせる。
# binsはビンの数（ヒストグラムの棒の数）。つまり多ければ細くなる。

plt.hist(dataset1,normed=True,alpha=0.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20, color='indianred')
```

```python
data1 = randn(1000)
data2 = randn(1000)

# 同時分布（結合分布）。散布図とヒストグラム
sns.jointplot(data1,data2)
```

```python
# もうすこし分かり易く
sns.jointplot(data1,data2,kind='hex')

# 色の濃さで分布が分かる
```
