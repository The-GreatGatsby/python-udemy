```python
# 乱数列の種を与えます。
np.random.seed(12345)
dframe = DataFrame(np.random.randn(1000,4))
```

```python
dframe.head()
```

```python
dframe.tail()
```

```python
# Lets describe the data
dframe.describe()
```

```python
# 最初の列
col = dframe[0]
```

```python
col.head()
```

```python
# 3より大きい要素を取り出します。
col[np.abs(col)>3]
```

```python
# DataFrameの全列に対して同じ演算をします。
dframe[(np.abs(dframe)>3).any(1)]
```

```python
np.sign(dframe)
```

```python
# 絶対値の上限が3になるようにします。
dframe[np.abs(dframe)>3] = np.sign(dframe) *3
```

```python
dframe.describe()
```
